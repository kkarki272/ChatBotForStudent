```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Load embedding model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings for FAQ questions
faq_embeddings = embedder.encode(faq_df['Question'].tolist(), convert_to_tensor=False)
faq_embeddings = np.array(faq_embeddings).astype("float32")

# Build FAISS index
index = faiss.IndexFlatL2(faq_embeddings.shape[1])
index.add(faq_embeddings)
```

```python
def get_answer(user_question):
    user_embedding = embedder.encode([user_question], convert_to_tensor=False)
    user_embedding = np.array(user_embedding).astype("float32")

    # Get top 3 matches
    D, I = index.search(user_embedding, k=3)

    best_score = D[0][0]
    best_index = I[0][0]

    # Confidence-based filtering
    if best_score > 3.5:
        return "I'm not sure I understand your question clearly. Could you please rephrase or ask something related to student services or campus info?"

    answer = faq_df['Answer'].iloc[best_index]
    return answer

```

```python
import gradio as gr

with gr.Blocks() as demo:
    gr.Markdown("# 🎓 Campus Support Chatbot")
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="Ask me anything about your campus")

    def respond(message, history):
        response = get_answer(message)
        history.append((message, response))
        return history, history

    msg.submit(respond, [msg, chatbot], [chatbot, chatbot])

demo.launch(share=True)
```
